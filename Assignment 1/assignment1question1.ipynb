# LOADING IRIS DATASET:-

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import datasets

iris=datasets.load_iris()
iris_dataframe=pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_dataframe.to_csv('iris_dataset.csv')
```

# PLOTTING HISTOGRAMS:-

```python
for column in iris_dataframe.columns[:]:
    bins=[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8]
    plt.hist(iris_dataframe[column],bins=bins)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Number of flowers')
    plt.xticks(bins)
    plt.show()
```

# PLOTTING SCATTER PLOTS:-

```python
def scatter_function(feature1,feature2):
    plt.scatter(iris_dataframe[feature1], iris_dataframe[feature2], c=iris.target,s=5) 
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.title(f'Scatter Plot: {feature1} vs {feature2}')
    plt.show()
scatter_function('sepal length (cm)', 'sepal width (cm)')
scatter_function('petal length (cm)', 'petal width (cm)')
scatter_function('sepal length (cm)', 'petal width (cm)')
scatter_function('petal length (cm)', 'sepal width (cm)')
scatter_function('petal length (cm)', 'sepal length (cm)')
scatter_function('petal width (cm)', 'sepal width (cm)')
```

# FINDING MEAN, MEDIAN, MODE AND STANDARD DEVIATION OF NUMERICAL COLUMNS:-

```python

print("The mean is :")
print(iris_dataframe.mean())
print("The median is :")
print(iris_dataframe.median())
print("The mode is :")
print(iris_dataframe.mode())
print("The standard deviation is :")
print(iris_dataframe.std())
```
